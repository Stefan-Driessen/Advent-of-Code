```python
class CRT:
    # Cathode class for day 10
    def __init__(self):
        self.X = 1
        self.cycle = 1
        
    def runCycle(self, val):
        self.X += val
        self.cycle += 1
        
    def signalStrength(self):
        return self.X * self.cycle
    
    def pixel(self):
        if self.cycle % 40 in [self.X, self.X + 1, self.X + 2]:
            return "#"
        else:
            return "."
    
def inputToVals(inp):
    # Take the input from the assignment and translates it to an array of changes.
    vals = []
    for i in inp.split('\n')[:-1]:
        if i == "noop":
            # At the end of this cycle, the X register has not changed.
            vals.append(0)
        else:
            # At the end of the next cycle, the X register changes via an addx
            vals = vals + [0, int(i[5:])]
    return vals
```

```python
with open('input.txt') as f:
    inputs = f.read()
    
vals = inputToVals(inputs)
```
