```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
```

```python
with open('input.txt') as f:
    rocks = f.read().split('\n')
```

```python
# 0 = 450
# 80 = 530
# 50 = 500
class Caves():
    def __init__(self, _rocks, depth=171, width=81, start=50):
        self.sandCounter = 0
        self.field = np.zeros([depth, width], dtype=str)
        self.field = np.full_like(self.field, '.')
        self.field[0][start] = '+'
        self.start = start
        self.width = width
        self.depth = depth

        for rock in rocks:
            initRockRcs = rock.split(' -> ')
            X, Y = initRockRcs[0].split(',')
            X, Y = int(X) - 500 + start, int(Y)
            self.field[Y][X] = '#'
            prevX, prevY = X, Y
            for rcs in rock.split(' -> '):
                X, Y = rcs.split(',')
                X, Y = int(X) - 500 + start, int(Y)
                if X < prevX:
                    for i in range(X + 1, prevX + 1):
                        assert Y == prevY, f"We've got some diagonal shit going from {X}, {Y} to {prevX}, {prevY}."
                        self.field[Y][i] = "#"
                elif X > prevX:
                    for i in range(prevX, X + 1):
                        assert Y == prevY, f"We've got some diagonal shit going from {X}, {Y} to {prevX}, {prevY}."
                        self.field[Y][i] = "#"
                elif Y < prevY:
                    for i in range(Y, prevY + 1):
                        assert X == prevX, f"We've got some diagonal shit going from {X}, {Y} to {prevX}, {prevY}."
                        self.field[i][X] = "#"
                elif Y > prevY:
                    for i in range(prevY, Y + 1):
                        assert X == prevX, f"We've got some diagonal shit going from {X}, {Y} to {prevX}, {prevY}."
                        self.field[i][X] = "#"

                prevX, prevY = X, Y
        
    def printCaves(self):
        for row in self.field:
            for val in row:
                print(val, end="")
            print()
        
    def dropSand(self):
        atRest = False
        curY, curX = (0, self.start)
        while not atRest:
            # print(curX, curY)
            # print(self.field[curY + 1][curX])
            # print(self.field[curY + 1][curX] == ".")
            # if curY > 169:
                # print(f'Going over the edge at {self.sandCounter}')
            if self.field[curY + 1][curX] == ".":
                # D
                curY += 1
            elif self.field[curY + 1][curX - 1] == ".":
                # DL
                curY += 1
                curX -= 1
            elif self.field[curY + 1][curX + 1] == ".":
                # DL
                curY += 1
                curX += 1
            else:
                self.field[curY][curX] = 'o'
                atRest = True
                self.sandCounter += 1

    def blockSand(self):
        count = 0
        for i, row in enumerate(self.field):
            for j, s in enumerate(row):
                if s not in ['.', '+'] :
                    count += 1
                    if (row[j-1] not in ['.', '+']) & (row[j+1] not in ['.', '+']) & (i < 171):
                        if self.field[i+1][j] == '.':
                            self.field[i+1][j] = 'X'
        return count

    def floor(self):
        for i, s in enumerate(self.field[-1]):
            self.field[-1][i] = '#'
```

```python
a = Caves(rocks)
```

```python
# a.dropSand()
# a.printCaves()
while True:
    a.dropSand()
```
