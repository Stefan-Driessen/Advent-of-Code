```python
with open('input.txt') as f:
    rows = f.read()
    
snackPacks = rows.split('\n\n')
```

```python
def cals(snackPack):
    cals = 0
    for snack in snackPack.split('\n'):
        cals += int(snack)
    return cals
```

```python
calories = []
for sp in snackPacks:
    calories.append(cals(sp))
calories.sort()
calories
```
