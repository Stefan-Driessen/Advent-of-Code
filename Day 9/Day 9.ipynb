```python
class Head():
    def __init__(self):
        self.X = 0
        self.Y = 0
        
    def __str__(self):
        return f"X: {self.X}, Y: {self.Y}"
        
    def move(self, d):
        if d == "L":
            self.X -= 1
        elif d == "R":
            self.X += 1
        elif d == "U":
            self.Y += 1
        else:
            assert d == "D", f'Got invalid direction for head: {d}'
            self.Y -= 1
            
class Tail(Head):
    def __init__(self):
        self.X = 0
        self.Y = 0
        self.visited = set((0, 0))
        
        def __str__(self):
            return f"X: {self.X}, Y: {self.Y}"
        
    def move(self, d):
        if d == "L":
            self.X -= 1
        elif d == "R":
            self.X += 1
        elif d == "U":
            self.Y += 1
        elif d == "D":
            self.Y -= 1
        elif d == "UL":
            self.X -= 1
            self.Y += 1
        elif d == "UR":
            self.X += 1
            self.Y += 1
        elif d == "DR":
            self.X += 1
            self.Y -= 1
        else:
            assert d == "DL", f'Got invalid direction for tail: {d}'
            self.X -= 1
            self.Y -= 1
        self.visited.add((self.X, self.Y))
        
class Rope():
    def __init__(self, H, T):
        self.head = H
        self.tail = T
        
    def move(self, inst):
        direction, size = inst.split(' ')
        size = int(size)
        for i in range(size):
            self.head.move(direction)
            if direction == 'L':
                if self.tail.X - self.head.X > 1:
                    assert self.tail.X - self.head.X == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.Y - self.tail.Y > 0:
                        assert self.head.Y - self.tail.Y == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UL')
                    elif self.tail.Y - self.head.Y > 0:
                        assert self.head.Y - self.tail.Y == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DL')
                    else:
                        self.tail.move('L')
            if direction == 'R':
                if self.head.X - self.tail.X > 1:
                    assert self.head.X - self.tail.X == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.Y - self.tail.Y > 0:
                        assert self.head.Y - self.tail.Y == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UR')
                    elif self.tail.Y - self.head.Y > 0:
                        assert self.head.Y - self.tail.Y == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DR')
                    else:
                        self.tail.move('R')
            if direction == 'U':
                if self.head.Y - self.tail.Y > 1:
                    assert self.head.Y - self.tail.Y == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.X - self.tail.X > 0:
                        assert self.head.X - self.tail.X == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UR')
                    elif self.tail.X - self.head.X > 0:
                        assert self.head.X - self.tail.X == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UL')
                    else:
                        self.tail.move('U')
            if direction == 'D':
                if self.tail.Y - self.head.Y > 1:
                    assert self.tail.Y - self.head.Y == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.X - self.tail.X > 0:
                        assert self.head.X - self.tail.X == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DR')
                    elif self.tail.X - self.head.X > 0:
                        assert self.head.X - self.tail.X == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DL')
                    else:
                        self.tail.move('D')
                        
class tailRope():
    def __init__(self, H, T):
        self.head = H
        self.tail = T
        
    def move(self, inst):
        direction, size = inst.split(' ')
        size = int(size)
        for i in range(size):
#             self.head.move(direction)
            if 'L' in direction:
                if self.tail.X - self.head.X > 1:
                    assert self.tail.X - self.head.X == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.Y - self.tail.Y > 0:
#                         assert self.head.Y - self.tail.Y == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UL')
                        return('UL')
                    elif self.tail.Y - self.head.Y > 0:
#                         assert self.head.Y - self.tail.Y == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DL')
                        return('DL')
                    else:
                        self.tail.move('L')
                        return('L')
            if 'R' in direction:
                if self.head.X - self.tail.X > 1:
                    assert self.head.X - self.tail.X == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.Y - self.tail.Y > 0:
#                         assert self.head.Y - self.tail.Y == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UR')
                        return('UR')
                    elif self.tail.Y - self.head.Y > 0:
#                         assert self.head.Y - self.tail.Y == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DR')
                        return('DR')
                    else:
                        self.tail.move('R')
                        return('R')
            if 'U' in direction:
                if self.head.Y - self.tail.Y > 1:
                    assert self.head.Y - self.tail.Y == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.X - self.tail.X > 0:
#                         assert self.head.X - self.tail.X == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UR')
                        return('UR')
                    elif self.tail.X - self.head.X > 0:
#                         assert self.head.X - self.tail.X == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('UL')
                        return('UL')
                    else:
                        self.tail.move('U')
                        return('U')
            if 'D' in direction:
                if self.tail.Y - self.head.Y > 1:
                    assert self.tail.Y - self.head.Y == 2, f"The tail {self.tail} is too far away from the head {self.head}"
                    if self.head.X - self.tail.X > 0:
#                         assert self.head.X - self.tail.X == 1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DR')
                        return('DR')
                    elif self.tail.X - self.head.X > 0:
#                         assert self.head.X - self.tail.X == -1, f"The tail {self.tail} is too far away from the head {self.head}"
                        self.tail.move('DL')
                        return('DL')
                    else:
                        self.tail.move('D')
                        return('D')
```
